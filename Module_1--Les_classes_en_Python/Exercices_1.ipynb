```python
#@title Exercice 1.2

class Client:
    def __init__(self, numero, nom, prenom, adresse=""):
        self._numero = numero
        self._nom = nom
        self._prenom = prenom
        self._adresse = adresse

    def afficher(self):
        message = f"#{self._numero}: {self._nom}, {self._prenom}\n\t{self._adresse}"
        print(message)

    # Pour l'exercice 1.4
    def get_numero(self):
        return self._numero

    def set_adresse(self, nouvelle_adresse):
        self._adresse = nouvelle_adresse

    
premier_client = Client(143, "Simpson", "Homer", "742 Evergreen Terrace, Springfield")
deuxieme_client = Client(875, "Flanders", "Ned", "744 Evergreen Terrace, Springfield")

premier_client.afficher()
print()
deuxieme_client.afficher()
```

```python
#@title Exercice 1.3

import csv

class Voiture:
    def __init__(self, marque, modele, emissions_co2):
        self._marque = marque
        self._modele = modele
        self._emissions_co2 = emissions_co2
    
    def get_marque(self):
        return self._marque
    
    def get_modele(self):
        return self._modele
        
    def get_emissions_co2(self):
        return self._emissions_co2

les_voitures = []   

# Lecture du fichier CSV
with open("voitures.csv", "r") as fichier_voiture:
    lecteur = csv.reader(fichier_voiture)
    entete = next(lecteur)
    indice_marque = entete.index("Marque")
    indice_modele = entete.index("Modele")
    indice_emissions = entete.index("EmissionsCO2")
    for ligne in lecteur:
        marque = ligne[indice_marque]
        modele = ligne[indice_modele]
        emissions = ligne[indice_emissions]
        # On crée l'instance directement avec les données qu'on retient
        les_voitures.append(Voiture(marque, modele, emissions))

# On tri en utilisant les émissions de CO2 comme clé de comparaison
les_voitures.sort(key=lambda v:v.get_emissions_co2())
for une_voiture in les_voitures:
    chaine = f"{une_voiture.get_marque()}, {une_voiture.get_modele()} ==> {une_voiture.get_emissions_co2()}"
    print(chaine)


```

```python
#@title Exercices 1.5 et 1.6

import numpy as np

class Vecteur2D:

    def __init__(self, x=0, y=0):
        self._x = x
        self._y = y

    def x(self):
        return self._x
 
    def y(self):
        return self._y

    def calculer_magnitude(self):
        return np.sqrt(self._x**2 + self._y**2)

    def __add__(self, other):
        nouveau_x = self._x + other._x
        nouveau_y = self._y + other._y
        return Vecteur2D(nouveau_x, nouveau_y)
    
    def __sub__(self, other):
        nouveau_x = self._x - other._x
        nouveau_y = self._y - other._y
        return Vecteur2D(nouveau_x, nouveau_y)
    
    def __neg__(self):
        nouveau_x = - self._x
        nouveau_y = - self._y
        return Vecteur2D(nouveau_x, nouveau_y)
    
    def __mul__(self, valeur):
        nouveau_x = valeur * self._x
        nouveau_y = valeur * self._y
        return Vecteur2D(nouveau_x, nouveau_y)
    
    # Pour l'exercice 1.5
    def __rmul__(self, valeur):
        return self * valeur

    def __str__(self):
        chaine = f"({self._x:.2f}, {self._y:.2f})"
        return chaine

vecteur_1 = Vecteur2D(2.0, 4.3)
vecteur_2 = Vecteur2D(4.5, 1.8)

vecteur_resultant = vecteur_1 + vecteur_2
magnitude = vecteur_resultant.calculer_magnitude()
print(f"Le vecteur {vecteur_resultant} a une magnitude de {magnitude:.2f}")

vecteur_direction = vecteur_2 - vecteur_1
print(f"La déplacement de {vecteur_1} à {vecteur_2} est {vecteur_direction}")
print(f"La déplacement de {vecteur_2} à {vecteur_1} est {-vecteur_direction}")
print(f"{vecteur_1} * 5 donne {vecteur_1 * 5}")

# Pour l'exercice 1.5
print(f"5 * {vecteur_1} donne {5 * vecteur_1}")
```
